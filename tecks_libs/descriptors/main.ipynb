# Descriptors

```python
# Sans @dataclasses decorator, il faut définir manuellement les méthodes __init__ et __repr__ pour que l'instance de Person puisse être affichée de manière lisible.

class Person:
    def __init__(self, name: str, age: int):
        self.name = name
        self.age = age

    def __repr__(self):
        return f"Person(name={self.name!r}, age={self.age!r})"


lio = Person("li", 21)
print(lio)
```

```python
import dataclasses

@dataclasses.dataclass
class Person:
  name: str
  age: int
  
lio = Person('li', 21)

print(lio)
```

Un descriptor = un objet qui contrôle un attribut

🎯 Quand utiliser un descriptor ?
* → valider automatiquement les valeurs
* → calculer un attribut à la volée
* → faire un champ “intelligent”
* → éviter de répéter du code dans plusieurs classes
* → créer un mini‑framework (ORM, datamodel, etc.)

```python
class Positive:
    def __set__(self, instance, value):
        if value < 0:
            raise ValueError("Doit être positif")
        instance.__dict__[self.name] = value

    def __set_name__(self, owner, name: str):
        self.name: str = name


class Person:
    age = Positive()


person = Person()
person.age = 10  # OK
# person.age = -5  # ❌ ValueError
print(f"{person.age = }")
```

```python
class Email:
    def __set__(self, instance, value):
        if "@" not in value:
            raise ValueError("Email invalide")
        instance.__dict__["email"] = value


class User:
    email = Email()


u = User()
u.email = "test@example.com"  # OK
# u.email = "pas un email"  # ❌ ValueError
print(f"{u.email = }")
```

Avec un typage poussé

```python
from typing import Any, Self


class Box:
    def __init__(self, value: Any):
        self.value = value

    def copy(self) -> Self:
        return type(self)(self.value)

    def alias(self) -> Self:
        return self

    def __repr__(self):
        return f"<Box(value = {self.value!r}, Instance id = {id(self)})>"

a = Box(21)
b= a.alias()
c = a.copy()
print('Source a', a, '\nalias  b', b,'\ncopy   c', c)
```
