# Dict Views

```python
scores: dict[str, int] = {
    "Alice": 85,
    "Bob": 92,
    "Charlie": 78
}

print(scores.keys())
print(scores.values())
print(scores.items())
print()    
for k, v in scores.items():
    print(f"{k}: {v} %")

```

```python
from collections.abc import ItemsView, KeysView, ValuesView

scores: dict[str, int] = {"Alice": 85, "Bob": 92, "Charlie": 78}

keys: KeysView[str] = scores.keys()
values: ValuesView[int] = scores.values()
items: ItemsView[str, int] = scores.items()

def show() -> None:
    print(keys)
    print(values)
    print(items)

show()

scores.update({"David": 88})
print()

show()
```
